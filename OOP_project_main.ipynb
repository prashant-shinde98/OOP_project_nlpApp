```python
import nlpcloud
class NLPApp:

  def __init__(self):

    self.__database = {}
    self.__first_menu()

  def __first_menu(self):

    first_input = input("""
    Hi! How would you like to proceed?
    1. Not a member? Register
    2. Already a member? Login
    3. Exit
    """)

    if first_input == '1':
      self.__register()
    elif first_input == '2':
      self.__login()
    else:
      exit()

  def __register(self):
      name = input('Enter name: ')
      email = input('Enter email: ')
      password = input('Enter password: ')

      if email in self.__database:
        print('Email already exits')
      else:
        self.__database[email] = [name,password]
        print('Registration successful')
        self.__login()


  def __login(self):
      email = input('Enter email: ')
      password = input('Enter password: ')

      if email in self.__database:
        if self.__database[email][1] == password:
          print('Login successful')
          self.__second_menu()
        else:
          print('wrong password')
          self.__login()

      else:
        print('This email is not registered')

  def __second_menu(self):

    second_input = input("""
    Hi! How would you like to proceed?
    1. NER
    2. language Detection
    3. Sentiment Analysis
    4. Grammer and spelling correcion
    5. summerization
    6. Intent classification
    7. Headline generation
    8. Logout
    """)

    if second_input == '1':
      self.__ner()
    elif second_input == '2':
      self.__language_detection()
    elif second_input == '3':
      self.__sentiment_analysis()
    elif second_input == '4':
      self.__gs_correction()
    elif second_input == '5':
      self.__summery()
    elif second_input == '6':
      self.__intent
    elif second_input == '7':
      self.__headline()
    elif second_input == '8':
      self.__login()
    else:
      exit()

  def __ner(self):
    para = input('enter the paragraph: ')
    search_term = input('What you would like to search')
    client = nlpcloud.Client("finetuned-llama-3-70b", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=True)
    response = client.entities(para, searched_entity=search_term)
    print(response)

  def __language_detection(self):
    para = input('Enter the text: ')

    client = nlpcloud.Client("python-langdetect", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=False)
    response = client.lang_detection(para)

  def __sentiment_analysis(self):
    para = input('Enter the text: ')

    client = nlpcloud.Client("distilbert-base-uncased-emotion", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=False)
    response = client.sentiment(para)

    L = []
    for i in response['scored_labels']:
      L.append(i['score'])

    index = sorted(list(enumerate(L)), key= lambda x:x[1], reverse=True)[0][0]

    print(response['scored_labels'][index]['label'])

  def __gs_correction(self):
    para = input('Enter the text: ')
    client = nlpcloud.Client("finetuned-llama-3-70b", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=True)
    response = client.gs_correction(para)
    print(response)

  def __summery(self):
    para = input('Enter the text: ')
    client = nlpcloud.Client("t5-base-en-generate-headline", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=False)
    response = client.summarization(para)
    print(response)

  def __intent(self):
    para = input('Enter the text: ')
    client = nlpcloud.Client("t5-base-en-generate-headline", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=False)
    response = client.summarization(para)
    print(response)


  def __headline(self):
    para = input('Enter the text: ')
    client = nlpcloud.Client("finetuned-llama-3-70b", "2b21ce3bf88d4ba196c69b7c42119a9cc8c33766", gpu=True)
    response = client.intent_classification(para)
    print(response)

obj = NLPApp()
```
